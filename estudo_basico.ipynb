```python
from random import randint
def calcula(x, a):
    while x != a:
        if x < a:
            x = int(input("Errou! O número é maior.Tente de novo: "))
        else:
            x = int(input("Errou! O número é menor.Tente de novo: "))
    return

a = randint(0,100)
x = int(input("Tente adivinhar um número de 1 a 100: "))
calcula(x, a)
print("Parabéns, você acertou! O número é {}.".format(a))
```

```python
def soma(lista):
    total = sum(lista)
    print("O total é %d"%(total))

qtd = int(input("Desejar somar quantos números? "))
lista = []
for i in range(qtd):
    numero = int(input("Entre com o %dº número: "%(i+1)))
    lista.append(numero)
soma(lista)
```

```python
class Carro:
    def __init__(self,numero_portas,preco): # Construtor 
        self.numero_portas = numero_portas
        self.preco = preco
        print("Carro instanciado com sucesso!")
    def numero_de_portas(self):
        return self.numero_portas
    def troca_de_portas(self,novo_numero_portas):
        self.novo_numero_portas = novo_numero_portas
carro_01 = Carro(4,70000)
print(carro_01.numero_portas) # Chama diretamente o atributo
print(carro_01.numero_de_portas()) #Chama o método
carro_01.troca_de_portas(5) # Chama a método para trocar o número de portas
print(carro_01.numero_portas)
```

```python
#HERANÇA
class Carro:
    def __init__ (self,fabricacao,quilometragem):
        self.fabricacao = fabricacao
        self.quilometragem = quilometragem
        marca_do_carro = "A definir"
        tipo = "Particular"
    def set_marca_do_carro(self,texto):
        self.marca_do_carro = texto
    def get_tempo_desde_fabricacao(self):
        tempo = 2020 - self.fabricacao
        return tempo
    def get_media_quilometragem(self):
        media = self.quilometragem / carro.get_tempo_desde_fabricacao()
        return media
    def set_marca_do_carro(self,texto):
        self.marca_do_carro = texto
    
class Taxi(Carro):#Herança   
    def __init__ (self,fabricacao,quilometragem,licenca):
        self.fabricacao = fabricacao
        self.quilometragem = quilometragem
        self.licenca = licenca

#Qual o tipo do veículo, particular ou táxi?
veiculo = input("Qual o tipo do carro? Particular (escreva P) ou Taxi (escreva T): ")    
while (veiculo != "P") and (veiculo != "p") and (veiculo != "T") and (veiculo != "t"):
    veiculo = input("Tipo inválido. Qual o tipo do carro? Particular (escreva P) ou Taxi (escreva T): ")  
if (veiculo == "P") or (veiculo == "p"):
    veiculo = "Particular"
else:
    veiculo = "Táxi"

#Se for táxi, a licença é própria ou alugada?
if (veiculo == "Táxi"):
    licenca = input("Qual a licença do táxi? Própria (escreva P) ou alugada (escreva A) ?")    
    while (licenca != "P") and (licenca != "p") and (licenca != "A") and (licenca != "a"):
        licenca = input("Licença inválida. Qual a licença do táxi? Própria (escreva P) ou alugada (escreva A) ?")
    if (licenca == "P") or (licenca == "p"):
        licenca = "própria"
    else:
        licenca = "alugada"

#Qual o ano de fabricação do veículo?
ano = int(input("Qual o ano do veículo? "))
while (ano < 1980) or (ano > 2020):
    ano = int(input("Ano inválido. Escreva um ano entre 1980 e 2020: "))

#Qual a quilometragem do veículo?
km = int(input("Qual o quilometragem veículo? "))
while (km < 0) or (km > 500000):
    km = int(input("Quilometragem inválida. Escreva uma quilometragem entre 0 e 500 mil: ")) 

#Qual a marca do veículo?
marca = input("Qual a marca/modelo do veículo? ")

print("")

if (veiculo == "Particular"):
    carro = Carro(ano,km)
    carro.set_marca_do_carro(marca)
    print("O carro é particular e o modelo é {}".format(carro.marca_do_carro))
    print("O carro tem {} anos desde sua fabricação.".format(carro.get_tempo_desde_fabricacao()))
    print("O carro rodou {} km em média por ano.".format(int(carro.get_media_quilometragem())))
    
else:
    taxi = Taxi(ano,km,licenca)
    taxi.set_marca_do_carro(marca)
    print("O carro é um táxi de licença {} e o modelo é {}".format(taxi.licenca,taxi.marca_do_carro))
    print("O táxi tem {} anos desde sua fabricação.".format(taxi.get_tempo_desde_fabricacao()))
    print("O táxi rodou {} km em média por ano.".format(int(taxi.get_media_quilometragem())))
```

```python
#ACESSAR OS ATRIBUTOS DA CLASSE PAI (RODAR A CÉLULA DE CIMA PRIMEIRO) 
class Bicicleta(Carro):   
    def __init__ (self,fabricacao):
        super().__init__ (self,fabricacao)
        self.fabricacao = fabricacao

bicicleta = Bicicleta(2015)#Passar o ano de fabricação da bicicleta
print("A bicicleta foi fabricada em {}".format(bicicleta.fabricacao))
```

```python
#ACESSAR OS MÉTODOS DA CLASSE PAI (RODAR DUAS CÉLULAS ACIMA PRIMEIRO)
class Notebook(Carro):
    def __init__ (self,fabricacao):#Construtor especial pois só precisa de um argumento
        self.fabricacao = fabricacao

    def get_tempo_de_uso(self):#Método do notebook que acessa o método do carro 
        return super().get_tempo_desde_fabricacao()

notebook = Notebook(2010)#Passar o ano de fabricação do notebook
print("O Notebook tem {} anos de uso".format(notebook.get_tempo_de_uso()))
```

```python
#POLIMORFISMO
class Homem():
    def __init__ (self):
        pass

    def get_descricao(self):
        self.descricao = "Eu sou homem"
        oqueeusou = self.descricao
        return oqueeusou

class Mulher(Homem):
    def __init__ (self):
        super().__init__ ()

    def get_descricao(self):
        self.descricao = "Eu sou mulher"
        oqueeusou = self.descricao
        return oqueeusou

#Ao instanciar um Homem o resultado é diferente de instanciar uma mulher
instancia = Homem()
#instancia = Mulher() 
print(instancia.get_descricao())
```
